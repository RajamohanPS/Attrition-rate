### Data description: <br>
The dataset contains 26 columns/factors and 1,470 rows/observations. <br>
The dataset consists of factors or reasons that can lead to attrition of employees <br>
Different columns are <br>
Age - Age of Employee <br>
BusinessTravel - Non-Travel, Travel_Rarely, Travel_Frequently <br>
Department - Human Resource, Research and Development and Sales <br>
DistanceFromHome - Office distance from Home of employees <br>
Education - 1- 'Below College', 2- 'College', 3- 'Bachelor', 4- 'Master', 5- 'Doctor' <br>
EducationField - Human Resource, Life Sciences, Medical, Marketing, Technical Degree, Others <br>
EnvironmentSatisfaction - 1 - 'Low', 2 - 'Medium', 3 - 'High', 4 - 'Very High' <br>
Gender Male or Female <br>
JobInvolvement - 1 - 'Low', 2 - 'Medium', 3 - 'High', 4 - 'Very High' <br>
JobRole - Human Resource. Sales Executive, Research Scientist, Laboratory Technician, Manufacturing Director, Healthcare Representative, Manager, Sales Representative, Research Director <br>
JobSatisfaction - 1 - 'Low', 2 - 'Medium', 3 - 'High', 4 - 'Very High' <br>
MaritalStatus - Single,Married,Divorced <br>
MonthlyIncome - Monthly income of employees in dollar value  <br>
NumCompaniesWorked - Number of companies where employee has worked <br>
OverTime - Yes or NO <br>
PercentSalaryHike - Percentage hike in the salary of employees <br>
PerformanceRating - 1 - 'Low', 2- 'Good', 3 - 'Excellent', 4- 'Outstanding' <br>
RelationshipSatisfaction - 1 - 'Low', 2 - 'Medium', 3 - 'High', 4 - 'Very High' <br>
TotalWorkingYears - Total experience of the employees <br>
WorkLifeBalance - 1- 'Bad' , 2 - 'Good', 3 - 'Better', 4 - 'Best' <br>
YearsAtCompany - Number of years since the employee has joined <br>
YearsInCurrentRole - Number of years in the current job role <br>
YearsWithCurrManager - Number of years with the current manager <br>

## Analysis:

#### Countplots of features:

```python
def count_feat(feat,df):
    plt.figure(figsize=(20,5))
    sns.countplot(x=feat,hue='Attrition',data=df)
    plt.show()
```

```python
for i in cat_ord_feat:
    count_feat(i,data)
    
```

We can get a better idea of the relation of the features and attrition by looking at the % of attrition for each feature's sub classes

```python
def attr_perc(feat,df):
    tru_attr = df[df['Attrition'] == 'Yes'].groupby(feat).count()['Attrition']
    fals_attr = df[df['Attrition'] == 'No'].groupby(feat).count()['Attrition']
    perc = tru_attr/(tru_attr+fals_attr)*100
    perc.plot(kind='bar')
    plt.show()
```

```python
def attr_perc_print(feat,df):
    tru_attr = df[df['Attrition'] == 'Yes'].groupby(feat).count()['Attrition']
    fals_attr = df[df['Attrition'] == 'No'].groupby(feat).count()['Attrition']
    perc = tru_attr/(tru_attr+fals_attr)*100
    print(perc)
```

```python
for i in cat_ord_feat:
    attr_perc_print(i,data)
```

#### Countplots of Attrition rates for each features:

```python
for i in cat_ord_feat:
    attr_perc(i,data)
```

From the above countplots of the Attrition rate, we can observe that:


BusinessTravel - Employees that travel frequently (\~24.9%) have the highest rate of attrition compared to the other two classes. <br>
Department - Employees in the Sales dept. (\~20.6%) have the highest rate of attrition, closely followed by the HR dept (\~19%). <br>
EducationField - Employees with HR background have the highest attrition rate (\~25.9%), closely followed by technical degrees and marketing. <br>
JobRole - Sales Representatives see the highest attrition rate (\~39.7%), followed by Lab technician (\~23.9%) and HR (\~23%). <br>
MaritialStatus - Single employees have the highest attrition rate (\~25.5%) among the other classes. <br>
OverTime - Employees who work over time have a higher attrition rate (\~30.5%).


JobInvolvement,JobSatisfaction,EnvironmentSatisfaction,Education,WorkLifeBalance - These all show a similar trend of higher attrition rates at lower level, and the rate decreases with higher levels. 

Gender,PerformanceRating - These features show similar attrition rates for both classes. <br>

### Building a model

Dropping PerformanceRating as the attrition rate is equal for both sub classes. Has little or no influence on Attrition Rate.

```python
data_d.drop(columns='PerformanceRating',axis=1,inplace=True)
data_d
```

Encoding:

```python
data_d=pd.get_dummies(data_d,columns=categorical)
data_d
```

```python
data_d.drop(columns=['OverTime_No','Gender_Male'],axis=1,inplace=True)
```

```python
Y = data_d['Attrition_Yes']
X = data_d.drop(columns='Attrition_Yes',axis=1,inplace=False)
X
```

Using SMOTE to balance the dataset

```python
from imblearn.over_sampling import SMOTE
sample = SMOTE()
X_os, Y_os = sample.fit_resample(X, Y)
X_os
```

```python
Y_os.value_counts()
```

We can see that the Attrition has been oversampled.

#### Train/ test splits

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_os, Y_os, test_size=0.25, random_state=2)
```

```python
print(len(X_train),len(X_test))
```

#### XGBoost classifier

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=100)
xgb.fit(X_train, Y_train)

Y_predict = xgb.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import confusion_matrix
c_m = confusion_matrix(Y_test, Y_predict)
print(c_m)
```

```python
from sklearn.metrics import accuracy_score, f1_score
accuracy_score(Y_test, Y_predict)
```

```python
f1_score(Y_test, Y_predict)
```
